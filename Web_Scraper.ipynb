# Extracting Posts and Authors

```python
post_class = soup.find('div', {'class': 'container posts'})
post_class
```

```python
all_posts = post_class.find('div', {'class': 'post-stream'})

#extract the leading post 
article_div = all_posts.find('article', {'id': "post_" + str(1)})
print('')
partial = article_div.find('div', {'class': 'names trigger-user-card'})
username = partial.find('span', {'class': "first username"}).getText()
print(username)
author_name = partial.find('span', {'class': "second full-name"}).find('a').getText()
print(author_name)
print('')

#extract the content 
content = article_div.find('div', {'class': 'cooked'}).getText()
print(content)
print('')

#extract date
date = article_div.find('div',{'class':"topic-map-post created-at"}).find('span').get('title')
print(date)
print('')

#extract linkes
likes = article_div.find('li',{'class':"secondary likes"}).find('span').getText()
print(likes)
```

```python
posts_dict = {}
posts_dict['username'] = username
posts_dict['author_name'] = author_name
posts_dict['content'] = content
posts_dict['date'] = date
posts_dict['likes'] = likes

posts_dict
```

```python
df = pd.DataFrame.from_dict(posts_dict,orient='index')
df
```

```python
df.to_csv('web_scraping_data.csv')
```

```python
post_links = []

# get initial height of the webpage
height = driver.execute_script("return document.body.scrollHeight")

# check whether you've reached the botton of the webpage
while True:
    # scroll down to the bottom of the webpage
    driver.execute_script("window.scrollTo(arguments[0], document.body.scrollHeight);", height)
    
    # parse HTML with beautiful soup
    soup = BeautifulSoup(driver.page_source, 'lxml')
    # locate the HTML sections with posts
    body = soup.find('body')
    main_outlet = body.find('div', {'class': "container list-container"})
    
    # get all hyperlinks of posts
    all_a = main_outlet.findAll('a', {'class': 'title raw-link raw-topic-link'})
    for a in all_a:
        post_link = a.get('href')
        post_links.append(post_link)
        
    # calculate the new height of the webpage
    newHeight = driver.execute_script("return document.body.scrollHeight")
    
    if newHeight == height:
        break
    
    height = newHeight
```
